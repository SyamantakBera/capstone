**Data Encoding using LabelEncoder**

We are using the **LabelEncoder** from the **sklearn.preprocessing** module to encode categorical columns in our dataset. The goal of data encoding is to convert categorical data into a numerical representation, making it easier for machine learning algorithms to process the data.

Categorical Columns:
We have identified three categorical columns in our dataset that require encoding:

**'Gender of the patient'**: Represents the gender of the patient.
**'City_Code'**: Corresponds to the city where the patient resides.
**'State_Code'**: Represents the state where the patient resides.
LabelEncoder Usage:

We create an instance of the LabelEncoder as label_encoder.
We iterate through each categorical column using a for loop.
For each column, we use label_encoder.fit_transform(df[column].astype(str)) to convert the categorical data into numeric encoded values.
We also store the unique values in the categorical column using unique_values = df[column].unique().
Then, we print the encoding information for each unique value along with its corresponding encoded value for that column.
Why Data Encoding:
Machine learning models typically work with numerical data, and categorical data needs to be transformed into numerical format for analysis and model training. LabelEncoder is a simple and effective technique to convert categorical data to numerical values. By encoding the categorical columns, we can use this transformed data to train machine learning models, making it easier for them to understand the relationships between different categories in the data.

**Please note that encoding is essential for machine learning models as they cannot directly process categorical data. The numerical representation helps the model interpret the data and learn patterns effectively, leading to accurate predictions and insights.**

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['Gender of the patient', 'City_Code', 'State_Code']
label_encoder = LabelEncoder()

for column in categorical_columns:
    encoded_values = label_encoder.fit_transform(df[column].astype(str))
    unique_values = df[column].unique()
    print(f"{column} Encoding:")
    for value, encoded_value in zip(unique_values, encoded_values):
        print(f"  {value}: {encoded_value}")
    print()

for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column].astype(str))

# Display the encoded and filled DataFrame
df.head()
```

Now, We will be importing the Dataset **"Status.xlsx"** Containing if a patient with certain ID has Liver disease or not. This is determined by their status value:<br>
**1: <font color=red>Implies they have Liver Disease**</font><br>
**2: <font color=green>Implies they have a healthy Liver**</font><br>

```python
status = pd.read_excel('Status.xlsx')
status.tail() #printing the tail of the status dataset
```

We now want to merge the **"df_filled"** dataset and **"Status"** to have a single dataframe using **Inner Join** with <font color= red>**ID**</font> as the common column.

1. Next, we merge the 'df' DataFrame with the 'status' DataFrame on the 'ID' column using the `pd.merge()` function. The `how='inner'` argument ensures that only the rows with matching 'ID' values in both DataFrames are included in the merged DataFrame.

2. The merged DataFrame is stored in the variable 'merged'.

By merging the two datasets, we combine information from both datasets based on the common 'ID' column. This allows us to analyze the relationship between the variables in the 'df' DataFrame and the 'Status' column from the 'status' DataFrame.


```python
merged = pd.merge(df, status, on='ID', how='inner')
merged.tail()
```

Now we will drop the ID Column and make a final dataframe called **"dataframe"**:

```python
dataframe = merged.drop('ID', axis=1)
dataframe.tail()
```

```python
print(dataframe.isnull().sum())
```

**Calculate the Mean of Each Column**

In this section of the code, we are calculating the mean value for each numerical column in the DataFrame df. The df.mean(numeric_only=True) method is used to compute the mean, excluding non-numeric columns.

**Fill Null Values with the Mean**

Next, we create a new DataFrame called df_filled, which is a copy of the original DataFrame df. We then use the **fillna()** method to replace any missing values (null or NaN) in the DataFrame with the corresponding **mean** value computed earlier. This operation ensures that all numerical columns now have their missing values replaced with the mean, making the data more complete and suitable for further analysis.

Overall, these steps help to handle missing data in the DataFrame by filling null values with the mean, which is a common strategy for dealing with missing numerical data.

**Note : Columns "City_Code","State_Code" have 0 empty values thus are unaffected by mean filling**

```python
# Calculate the mean of each column
mean_values = dataframe.mean(numeric_only=True)

# Fill null values with the mean
dataframe = dataframe.fillna(mean_values)
dataframe.head()
```

```python
print(dataframe.isnull().sum())
```

#  <span style="color:navy"><center>VISUALISATION</center></span>

To create data visualizations in Python, we'll be using the **`matplotlib`** library along with the **`seaborn`** library, which provides a high-level interface for creating attractive and informative statistical graphics. 

We can import these libraries with the following lines of code:


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

In this section, we will group the data based on the patients' age and their liver disease status. Then, we will create a bar plot to visualize the distribution of liver disease status across different age.

First, let's group the data using the `groupby` function and calculate the count of patients with each status for each age group. We will store this data in a DataFrame called `age_groups`.


```python
# Grouping the data
age_groups = dataframe.groupby('Age of the patient')['Status'].value_counts().unstack()

# Creating the bar plot with increased size
age_groups.plot(kind='bar', stacked=True, figsize=(20, 8))

# Set x and y axis labels and title
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Status across Age Groups')

# Legend Titles
plt.legend(labels=['Liver Disease', 'No Liver Disease'], title='Status')

# Show the plot
plt.show()
```

```python

# Group the data and calculate the percentage of patients with status 1 for each age
age_groups = dataframe.groupby('Age of the patient')['Status'].value_counts().unstack()
age_groups['Percentage Status 1'] = (age_groups[1] / (age_groups[1] + age_groups[2])) * 100

# Set a modern style using Seaborn
sns.set_style('whitegrid')

# Increase the plot size
plt.figure(figsize=(20, 6))

# Create the bar plot using Seaborn
sns.barplot(x=age_groups.index, y='Percentage Status 1', data=age_groups, color='skyblue')

plt.xlabel('Age')
plt.ylabel('Percentage of Status 1')
plt.title('Percentage of Liver Patients for Each Age')

plt.show()
```

```python
# Group the age into 10-year bins
age_groups = pd.cut(dataframe['Age of the patient'], bins=range(0, 101, 10), right=False)

# Calculate the percentage of status 1 for each age group
grouped_percentage = dataframe.loc[dataframe['Status'] == 1].groupby(age_groups)['Status'].count() / dataframe.groupby(age_groups)['Status'].count() * 100

# Increase the plot size
plt.figure(figsize=(10, 6))

# Create the bar plot
plt.bar(grouped_percentage.index.astype(str), grouped_percentage)

# Set the x-axis label
plt.xlabel('Age Group')

# Set the y-axis label
plt.ylabel('Percentage of Liver Patient')

# Set the title
plt.title('Percentage of Liver Patients for Each Age Group')

# Show the plot
plt.show()
```

```python
# Filter the dataset for 'Status' 1 (Liver disease patients) and remove gender values of 2(Null)
status_1_df = dataframe[(dataframe['Status'] == 1) & (dataframe['Gender of the patient'] != 2)]

# Group the data by 'Age of the patient' and 'Gender of the patient' and count the occurrences
age_gender_counts = status_1_df.groupby(['Age of the patient', 'Gender of the patient']).size().reset_index(name='Count')

# Create a bar plot for age vs gender
plt.figure(figsize=(20, 6))
sns.barplot(data=age_gender_counts, x='Age of the patient', y='Count', hue='Gender of the patient', palette='Set1')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age vs Gender of Liver Disease Patients (Status 1)')
plt.legend(title='Gender', loc='upper right')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Separate the dependent variable (Status) and independent variables (all other columns)
X = dataframe.drop('Status', axis=1)
y = dataframe['Status']

# Add a constant column to the independent variables
X = sm.add_constant(X)

# Fit the multivariate regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

```python

from sklearn.svm import SVR
from sklearn.feature_selection import SelectKBest, f_regression

# Separate the dependent variable (Status) and independent variables (all other columns)
X = dataframe.drop('Status', axis=1)
y = dataframe['Status']

# Perform feature selection to select the top K features
k = 5  # Select the top 5 features
selector = SelectKBest(score_func=f_regression, k=k)
X_selected = selector.fit_transform(X, y)

# Create and fit the SVR model
model = SVR()
model.fit(X_selected, y)

# Get the feature scores and their corresponding column names
feature_scores = selector.scores_
feature_names = X.columns[selector.get_support()]

# Print the feature scores and their contribution in predicting the 'Status' variable
for feature, score in zip(feature_names, feature_scores):
    print(f"Feature: {feature}, Score: {score}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Separate the dependent variable (Status) and independent variables (all other columns)
X = dataframe.drop(['Status'], axis=1)
y = dataframe['Status']

# Initialize the RandomForestClassifier
clf = RandomForestClassifier(random_state=42)

# Fit the classifier on the data
clf.fit(X, y)

# Get feature importances from the classifier
importances = clf.feature_importances_

# Create a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importance_df.sort_values(by='Importance', ascending=False, inplace=True)

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Status 1')
plt.show()
```
